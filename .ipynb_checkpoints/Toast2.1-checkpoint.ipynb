## The Problem

The basic problem is this. When toast teeters off the edge of a standard counter it seems to, remarkably consistently, land 'jelly side' down. I have read that this is no accident and that in fact the solution to the problem is to push the toast when you notice it going off the edge rather than try to stop it. I have done some loose experiments and find this to be empirically true. Now -- can I use basic dynamics and an numerical approach to explore this problem without getting caught up the analytic solution to the problem documented in various places in AJP. 

In the previous notebook I modelled this process assuming that the angular acceleration would be constant. I know this is incorrect but I wanted to see if it gave me reasonable numbers. Experimentally starting with a piece of toast 3/4 of the way off the edge and then releasing it I observe that it rotates about $2\pi$ radians before it hits the floor (a little more perhaps). My previous model predicted 10ish radians which is more than $3\pi$ radians but it seems likely that the simple average model overestimates the angular velocity as the toast comes off the edge. 

In this model I will recalculate the torque and angular acceleration at small increments of time (an iterative or finite element model).

A quick calculation I should do as soon as these constants are established is calculate the time it take for the toast to fall from the counterheight to the floor. This is essentially a calculation we have done in lab and is given by some form of kinematic analysis and yields

$time_{fall} = \sqrt{\frac{2*h}{g}}$

From a standard counter this should yield around 0.45 s. As I go along I check each number that results against a quick manual calculation to check moth units and magnitude.


````python
tlength = 10.0
twidth = 10.0
tthick = 1.0
tdensity = 0.45
counterheight = 100.0
latspeed = 0.0
gravity = 981.0
anglimit = np.pi/2  # np.pi accesses the value of pi
overhang = .75

floortime = np.sqrt([2.0*counterheight/gravity])
print("The time to fall to the floor from the counter is %.4f seconds" % (floortime))
````

## Moment of Inertia (constant for this model)

In this case I want to start out by calculating the moment of inertia of the toast around it's center of mass in appropriate units. Because I am dealing with the toast rotating around the same point at all times the moment of inertia is constant.

Toast is essentially just a book (rectangular box) so it's moment of intertia is   
  
$I_{toast} = \frac{1}{12}m a^2$  

where a is the side perpendicular to the axis of rotation. The mass is calculated from the linear density and the volume of the toast.

### Parallel Axis:

We calculate the moment of inertia around an axis parallel to the axis through the center of mass using the Parallel Axis Theorem (brilliant eh?).

I$_{\parallel} = $I$_{cm} + m d^2$ 

where m is the total mass of the object and d is the perpendicular distance between the desired axis and the parallel one through the center of mass.

Just to test the calculation....we expect the moment of inertia to go up as we move away from the center of mass!

````python
lindensity = tdensity*tlength*tthick  # linear density of toast
tmass = lindensity*twidth # mass of toast

tinertiacm = tmass*(twidth*twidth)/12.0 # moment of inertia around CM
dparallel = twidth*(overhang - .50) # Overhang - half the width of the toast is the distance to axis of rotation.
tparallelI = tinertiacm + tmass*dparallel**2

# debug
# print ("Toast mass (gr): %.2f gr"% (tmass))
# print ("Inertia around CM: %.2f gr*cm^2" % (tinertiacm))
# print ("Inertia around parallel axis is %.2f gr*cm^2 at a distance of %.2f cm from CM" % (tparallelI, dparallel))
````

## FEA/Numerical Analysis/Interative Models

You have actually been introduced to this concept in a number of places both in math and physics. In fact, I would argue, this approach is most valuable in engineering where the truly interesting problems are so complex that there may be no hope of an analytic (mathematical expression) solution. Adding up rectangles in calculus to get the area under the curve is ultimately replaced by the formal process of integration and the antierivative. Nevertheless it is sometimes still the most effective strategy to just add up the areas of skinny rectangles.

## Iteration

Here is the conceptual frame. I calculate the initial angular acceleration. I then let a small amount of time elapse and then I figure out how fast the toast is moving and how far it has rotated. At the end of that small amount of time I recalculate the new angular acceleration not that the toast has rotated. Then I let the same small amount of time elapse and repeat the calculation for velocity and position. This is one interation. I repeat this process until I see the toast has rotated to the vertical and there is no more angular acceleration. I repeat the calculation to see how far it rotates at constant angular velocity unti it hits the floor. 

## How small is small enough?

This is in fact a very important question. The cool thing about computer code is that if I build it right I can make the small increment of time larger or smaller and see what happens to the result. When the results seem to converge and stabilize then that gives me a sense for the result. People have done PhD work looking at bad things that can happen as a result of unfortunate choices of step size (increment). In the begining we will take the optimistic approach that the trolls will stay out of sight and we won't fall into any traps.



## Iteration Setup

I'm going to need to keep track of angular position, velocity, and acceleration at each iteration. In Excel you would create a table to keep track of these values. In Python we create arrays (lists of numbers with the same name). [Numpy Arrays](https://docs.scipy.org/doc/numpy-1.15.0/user/basics.creation.html)

In this case I don't really have a good sense of how long it will take for the toast to rotate $\pi/2$ radians but it seems like it will be a fraction of a second. I will define a variable called

tstep (in ms) -- need to be a floating point number like 2.0 for calculational purposes

which represents the amount of time for each step and

numit (maximum number of iterations) -- this need to be an integer

This means my arrays need to have 'numit' elements and that's how I will establish them. In many coding languages we can't create arrays as we go. The need to be defined up front so the software can set aside space for them.

[np.linspace](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html)  
[np.full_like](https://docs.scipy.org/doc/numpy/reference/generated/numpy.full_like.html#numpy.full_like)

Are the ones that have been most useful to me.

```python
count = np.linspace(0,numit,num=numit+1) # start at 0, go to numit, because it started at there is 1 more element in the array
currenttime = np.full_like(count,0.0) # same size as count will all values = 0 for starters
angaccel = np.full_like(count,0.0)
angvel = np.full_like(count,0.0)
angpos = np.full_like(count,0.0)
torqpos = np.full_like(count,0.0)
torqneg = np.full_like(count,0.0)
torqtot = np.full_like(count,0.0)
```
Now that I am creating a new notebook for each improvement to the model the variable names will be unique within this notebook.

````python
tstep = 10.0 # in ms
numit = 20  # max number of iterations if toast doesn't rotate pi/2

count = np.linspace(0,numit,num=numit+1) # start at 0, go to numit, because it started at there is 1 more element in the array

# the various array variables
currenttime = np.full_like(count,0.0) # same size as count will all values = 0 for starters
angaccel = np.full_like(count,0.0)
angvel = np.full_like(count,0.0)
angpos = np.full_like(count,0.0)
torqpos = np.full_like(count,0.0)
torqneg = np.full_like(count,0.0)
torqtot = np.full_like(count,0.0)
````

## Filling the currenttime array

I intend for the current time array to contain that actual elapsed time for future plotting use. It also provides a useful introduction to loops. The 'while' loop has been a good starting point for many of my projects. 

[while](http://anh.cs.luc.edu/handsonPythonTutorial/whilestatements.html)  

Something that is painful is being sure that you know what the last value in the array is. In my first test I set numit = 50 and made the count array go from 0 to 50 which is 51 numbers. While it is truly 51 numbers the last value in the list is currenttime[50] NOT currenttime [51]. It's just something that everyone has to get used to with the way computers track addresses.



````python
ndex1 = 0 # this is an index to track how many times I've done the loop
while ndex1 < (numit+1):
    currenttime[ndex1] = float(ndex1*tstep)
    ndex1 = ndex1 + 1

# debug
# print("5th time element is %.3f ms. " % (currenttime[5]))
# print("Last time element is %.3f ms. " % (currenttime[numit]))
````

## Calculate the acceleration...

No - wait the torque is going to change with every interation so I need an array for that. This is how it actually happened so I wanted to capture that experience for you:) A really nice thing about Jupyter notebooks (that can also be a problem) is that when I forget something I go back just to that point, make the modification, and then run all subsequent cells until I get back to where I was.

As I was writing this relatively complicated calculation I keep finding a need to change my initial strategy. In calculating the positive and negative torques I didn't previously account for the possible angular rotation in determining the moment arm. 

```python
 torqpos[ndex2] = (overhang*twidth/2.0)*(tmass*overhang*gravity)
```
The second term (the force) doesn't change. Because of the rotation I need to reduce the moment arm by the cos(angular position) which is stored in the angpos[] array. I am calculating all angles in radians which is what the trig functions in Python expect. If I change it to:

```python
 torqpos[ndex2] = np.cos(angpos[ndex2]*(overhang*twidth/2.0)*(tmass*overhang*gravity)
```
that will take care of the rotation.

At this point I'm realizing that I going to need to check what is going on as I develop this calculation. I wanted to step each time through the 'while' loop and see what the variables are doing. If things seem out of whack I can stop the execution (the square next to the run command on the Jupyter notebook). There are other ways that professional coders use when working in their IDE of choice (Spyder, PyCharm etc) but this is a way to accomplish the goal albeit in a slightly clunky method. When I'm done I will comment this out.

```python
    # debugging help
    # print("Tpos: %.3f Tneg %.3f Ttot %.3f angaccel %.3f " % (torqpos[ndex2],torqneg[ndex2],torqtot[ndex2],angaccel[ndex2]))
    print("pos %.3f pos+ %.3f vel %.3f vel+ %.3f accel %.3f " % (angpos[ndex2],angpos[ndex2+1],angvel[ndex2],angvel[ndex2+1],angaccel[ndex2]))

```


As before I need to now calculate the angular position of the toast after 1 time step (tstep) similarly to how I did it before. I should know the initial velocity and position which allows me to calculate the position and velocity at the end of the increment of time. In doing this I had to be careful to account for the tstep being in ms (first try gave ridiculous numbers! -- see debugging above)

An important note: Notice how the last two lines are indented differently than the code within the loop. Indents actually mean things to the python code. After the toast reaches $\pi/2$ and I break out of the loop I printed out the final state of the toast before it comes off the edge for reference. I am about to plot these variables which will make it less relevant but it makes me feel good.



````python
ndex2 = 0 # a different tracking index
while (ndex2 < numit) and (angpos[ndex2], anglimit):
    # print("iteration: ",ndex2)
    
    # calculate current torques
    torqpos[ndex2] = np.cos(angpos[ndex2])*(overhang*twidth/2.0)*(tmass*overhang*gravity)
    torqneg[ndex2] = -(np.cos(angpos[ndex2])*(1.0-overhang)*twidth/2)*(tmass*(1.0-overhang)*gravity)
    torqtot[ndex2] = torqpos[ndex2]+torqneg[ndex2]
    
    # now calculate the angular acceleration
    angaccel[ndex2] = torqtot[ndex2]/tparallelI
    
    # determine next position and velocity after tstep                  
    angvel[ndex2+1] = angvel[ndex2] + angaccel[ndex2]*(tstep/1000.0)
    angpos[ndex2+1] = angpos[ndex2] + angvel[ndex2]*(tstep/1000.0) + 0.5*angaccel[ndex2]*(tstep/1000.0)*(tstep/1000.0)
    
    # debugging help
    # print("Tpos: %.3f Tneg %.3f Ttot %.3f angaccel %.3f " % (torqpos[ndex2],torqneg[ndex2],torqtot[ndex2],angaccel[ndex2]))
    # print("cos(angpos): ", np.cos(angpos[ndex2]))
    # print("pos %.3f pos+ %.3f vel %.3f vel+ %.3f accel %.3f " % (angpos[ndex2],angpos[ndex2+1],angvel[ndex2],angvel[ndex2+1],angaccel[ndex2]))
    # inputcont = input("continue?")    
    
    # test for end point of rotation   
    if angpos[ndex2+1] > (np.pi/2.0):
        ndex2 = ndex2 + 1
        print ("Got to 90 degrees at ndex2: ", ndex2)
        break # get out of the loop
        
    ndex2 = ndex2 +1 # go to the next time increment
    
# debug
print("Ttot %.3f angaccel %.3f : Both should be 0.0" % (torqtot[ndex2],angaccel[ndex2]))
print("pos %.3f vel %.3f : position = 1.55ish, final velocity as toast leaves edge" % (angpos[ndex2],angvel[ndex2]))
````

 ## Calculate Rotation
 
 Repeat the calculation in the previous notebook.....

````python
angposflr = angvel[ndex2]*floortime

print("Rotation at floor: %.3f radians. " % (angposflr))
````
