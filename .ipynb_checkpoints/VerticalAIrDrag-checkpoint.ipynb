# An 1D Example of Air Drag

This follows after an iteration assignment where the trajectory (x,y) of a ball was calculated iteratively. Because all the forces are constant the iterative model for such an ideal physics ball exactly follows the kinematic equations. This notebook explores what happens if the forces change (although I'm not including a change in direction in this model)

### Vertical Motion

This model represents a ball thrown straight up with air drag included. I will endeavor to track both the ideal ball and the ball with air drag to try to capture the differences. The drag model I am using is:

$\large F_{drag} = C*v^2$

I recognize that the drag force actually also depends on the density of the air and the cross section of the object. [see [`Air Drag`](https://en.wikipedia.org/wiki/Drag_(physics))] but I'm going to take all those other factors to be constant and rolled into my drag coefficient.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.rcParams["figure.figsize"] = (20,20)
plt.rcParams.update({'font.size': 22})
import numpy as np
```

### Constants

Establishing the basic parameters of the model.

initvel - initial upward velocity in m/s    
mass (kg)  
gravity - standard gravity in m/s/s  
dragcoef - drag coefficient unitless 
  
numiterations (integer)  
timestep = (ms)  
dtime (s) - calculated from timestep to get seconds  
  
pi - just easier than np.pi


```python
initvel = 5.
mass = 5.
gravity = -9.81
dragcoef = 0.05

numiterations = 50
timestep = 100.
dtime = timestep/1000.0


pi = np.pi
```

### Set up Arrays

These arrays represent all the characteristics of the motion of the object that I intend to keep track of. I don't know if it's wise but for now I tend to set up all the arrays with 0's in every location. 


ypos - initially 0, ideal physics ball (m)  
yposdrag - initially 0, air drag (m)  
yvel - upward velocity m/s ideal - initial upward velocity chosen by user  
yveldrag - yvel as a result of drag forces   
dragforce - added effect of air drag - gravity assumed constant (N)  
yacceldrag - combination of gravity and drag force  


```python
count = count = np.linspace(0,numiterations-1,num=numiterations) # start at 0, go to numiterations
currenttime = np.full_like(count,0.0) # same size as count will all values = 0 for starters

ypos = np.full_like(count,0.0)
yposdrag = np.full_like(count,0.0)
yvel = np.full_like(count,0.0)
yveldrag = np.full_like(count,0.0)
dragforce = np.full_like(count,0.0)
yacceldrag = np.full_like(count,0.0)
```

### Adjust initial conditions of arrays

ypos - initially 0, already set  
yposdrag - initially 0, already set  
yvel - initial velocity  
yveldrag - initial velocity  
dragforce - calculated initially  
yacceldrag - calculated initially   



```python
yvel[0] = 25.
yveldrag[0] = yvel[0]
dragforce[0] = -dragcoef*yveldrag[0]**2
yacceldrag[0] = dragforce[0]/mass + gravity

print("Initial upward velocity:", yveldrag[0])
print("Initial drag force (N):", dragforce[0])
print("Initial y acceleration with drag:", yacceldrag[0])
```

### Calculate next position and velocity from current one.

Using the kinematic expressions developed in PH211. I will assume that the acceleration is constant in direction and magnitude for the duration of an increment of time. Knowing current position and velocity I can then determine the position and velocity a 'short' time later using:

$\large y_{next} = y_{current} + v_{y_{current}}*t  + \frac{1}{2}*a_{y_{current}} * t^2$  
$\large v_{y_{next}} = v_{y_{current}} + a_{y_{current}}* t$

...and similarly for the arrays with drag



```python
ndex1 = 0

while ndex1 < numiterations-1:
    # time
    currenttime[ndex1+1] = currenttime[ndex1]+timestep
    # positions
    ypos[ndex1+1] = ypos[ndex1] + yvel[ndex1]*dtime + 0.5*gravity*dtime**2
    yposdrag[ndex1+1] = yposdrag[ndex1] + yveldrag[ndex1]*dtime + 0.5*yacceldrag[ndex1]*dtime**2
    # velocities
    yvel[ndex1+1] = yvel[ndex1] + gravity*dtime
    yveldrag[ndex1+1] = yveldrag[ndex1] + yacceldrag[ndex1]*dtime
    
    # test for hit ground 
    if ypos[ndex1+1] < 0.:
        ndex1 = ndex1 + 1
        print ("hit ground at t (ms): ", currenttime[ndex1])
        break # get out of the loop
                
    # drag force and acceleration
    dragforce[ndex1+1] = -dragcoef*yveldrag[ndex1]**2
    yacceldrag[ndex1+1] = dragforce[ndex1+1]/mass + gravity
    
    # Debug info
    # print("ndex1: ", ndex1)
    print("ypos: %.3f yposdrag %.3f : " % (ypos[ndex1+1],yposdrag[ndex1+1]))
    print("yvel: %.3f yveldrag %.3f : " % (yvel[ndex1+1],yveldrag[ndex1+1]))
    # print("dragforce: %.3f yacceldrag %.3f : " % (dragforce[ndex1+1],yacceldrag[ndex1+1]))
    # inputcont = input("continue?")  
    
    ndex1 = ndex1+1
    
print("Done with calculations!!")
```

### Plot the path

Having generated all this data it seems prudent to plot the path to see if the results make sense. We will explore the changes in the path that result from different choices of timestep. This will help illustrate a key feature/challenge of iterative models.

What I've created is a simple scatter plot of the x,y position of the object at each moment in time. The notebook version of the plot has some nice interactive features that aren't there in the inline version (see note about magic commands at begining of notebook)

```python
plt.scatter(currenttime, ypos, marker='.', color = 'green');
```

## Dual Plot

This is a way to plot two sets of data with a shared horizontal axis (time in this case). This is not particularly satisfying because the vertical axis is scaled differently in each case which makes it hard to understand what I'm seeing. I need to do this in a way which shares common horizontal and vertical axes.

```python
fig, yposplt = plt.subplots()

color = 'tab:red'
yposplt.set_xlabel('x (m)')
yposplt.set_ylabel('y (m)', color=color)
yposplt.plot(currenttime, ypos, color=color, marker = '.', label = 'ideal position')
yposplt.tick_params(axis='y', labelcolor=color)
plt.legend(loc = 'upper left')

ydragplt = yposplt.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ydragplt.set_ylabel('y (m)', color=color)  # we already handled the x-label with ax1
ydragplt.plot(currenttime, yposdrag, color=color, label = 'drag position')
ydragplt.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.legend(loc = 'upper right')
plt.show()
```

## Dual Plot

This approach certainly works but I think there's a better way. For now I've run out of time so I'm going to live with it but I will be updating this to be better at some point in the future.

```python
plt.plot(currenttime, ypos, marker = ".", color = "blue", label = "ideal")
plt.plot(currenttime, yposdrag, marker = ".", color = "red", label = "drag")
plt.title("Ideal and Drag")
plt.ylabel("position (m)");
plt.legend();
```
