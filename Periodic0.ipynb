```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
count = 1

from random import seed
from random import random
from random import randint
# seed random number generator
seed()
```

Since the set of plausible periodic motion problems is relatively circumscribed this next section is where I will seek to establish variable conventions for typical quantities.

```python
amplitude = 1.0 # in arbitrary units
period = 2.0  # also in arbitrary units
phaseshift = 0.0 # radians

xlabel = ["time (ms)", "time (s)", "time (min)"]
ylabel = ["position (m)", "velocity (m/s)","acceleration (m/s/s)"]
```

If I have the students start with this next cell to generate new practice problems then it will keep generating new cases for them.

```python
# generate practice cases

datatype = randint(0,2) # chooses what the periodic function is describing (position, velocity, acceleration)
periodtype = randint(0,2) # chooses the timescale (ms, s, min)

baseamp = random()
baseperiod = random()
basephase = random()

scaleamp = randint(1,21)
scaleperiod = randint(1,100)
scalephase = randint(1,6)  # 6 is an integer stand in for 2pi

amplitude = baseamp*scaleamp
period = baseperiod*scaleperiod
phase = basephase*scalephase

print("datatype: ", datatype)
print("periodtype: ", periodtype)
print("Amplitude: ",baseamp, scaleamp, amplitude)
print("Period: ",baseperiod, scaleperiod, period)
print("Phase (rad): ",basephase, scalephase, phase)
```

```python
x = np.linspace(0, 4*period, 1000)

plt.plot(x, np.sin(x))
plt.title("Periodic Behavior")
plt.xlabel(xlabel[periodtype])
plt.ylabel(ylabel[datatype]);
```

The position, size, and style of these labels can be adjusted using optional arguments to the function. For more information, see the Matplotlib documentation and the docstrings of each of these functions.

When multiple lines are being shown within a single axes, it can be useful to create a plot legend that labels each line type. Again, Matplotlib has a built-in way of quickly creating such a legend. It is done via the (you guessed it) `plt.legend()` method. Though there are several valid ways of using this, I find it easiest to specify the label of each line using the `label` keyword of the plot function:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

As you can see, the `plt.legend()` function keeps track of the line style and color, and matches these with the correct label. More information on specifying and formatting plot legends can be found in the `plt.legend` docstring; additionally, we will cover some more advanced legend options in [Customizing Plot Legends](https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html).
