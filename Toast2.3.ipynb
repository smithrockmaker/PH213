````python
# Define variable arrays needed
# time variables
count = np.linspace(0,numit,num=numit+1) # start at 0, go to numit, because it started at there is 1 more element in the array
currenttime = np.full_like(count,0.0) # same size as count will all values = 0 for starters

# range of parameters
coeffric = np.linspace(0.2,1.0,5) # let the coefficient of friction vary from 0.2 to 1.0 by 0.2
overhang = np.linspace(0.55,0.95,5) # initial overhang varies from .55 to .95 in percentage terms

# set up integers for generating multidimensional arrays....
numcoef = coeffric.size
numover = overhang.size
numcount = count.size

# moment of inertia variables
dparallel = np.full((numcoef,numover,numcount),0.) # distance to pivot from center of mass (CM)
tinertianow = np.full((numcoef,numover,numcount),0.) # moment of inertia from parallel axis theorem

# rotation variables
angaccel = np.full((numcoef,numover,numcount),0.) # current angular acceleration
angvel = np.full((numcoef,numover,numcount),0.) # current angular velocity
angpos = np.full((numcoef,numover,numcount),0.) # current angular position
torqpos = np.full((numcoef,numover,numcount),0.) # torque from overhanging 'right' side of toast
torqneg = np.full((numcoef,numover,numcount),0.) # torque from 'left' side of toast still over the table
torqnet = np.full((numcoef,numover,numcount),0.) # net torque

# general location of cm variables
rside = np.full((numcoef,numover,numcount),0.) # length of toast hanging out over edge
lside = np.full((numcoef,numover,numcount),0.) # length of toast to left of edge

# torque calculation variables
armr = np.full((numcoef,numover,numcount),0.)  # moment arm of overhanging toast
arml = np.full((numcoef,numover,numcount),0.)  # moment arm of toast left of pivot
weightr = np.full((numcoef,numover,numcount),0.)  # weight of overhanging toast acting at armr/2
weightl = np.full((numcoef,numover,numcount),0.)  # weight of 'left' side of toast acting at arml/2

# slipping variables
friction = np.full((numcoef,numover,numcount),0.) # friction at pivot
latgforce = np.full((numcoef,numover,numcount),0.) # force seeking to slide toast off
parallelaccel = np.full((numcoef,numover,numcount),0.) #  acceleration parallel to plane of toast
ndexfinal = np.full((numcoef,numover),0) # index of moment when toast leaves edge - note INTEGER!

# These arrays had to be added later as I needed to deal with the toast slipping off the edge
slipdisplace = np.full((numcoef,numover,numcount),0.) # displacement of toast in this interation
slipposx = np.full((numcoef,numover,numcount),0.) # position of CM of toast in x
slipposy = np.full((numcoef,numover,numcount),0.) # position of CM of toast in y
slipveltot = np.full((numcoef,numover,numcount),0.) # total velocity at iteration
slipvelx = np.full((numcoef,numover,numcount),0.) # velocity of CM in x direction
slipvely = np.full((numcoef,numover,numcount),0.) # velocity of CM in y direction
slipangle = np.full((numcoef,numover),0.) # angle at which slip begins

# kinematic coefficients 
quadcoef = np.full((numcoef,numover,3),0.) # needed to invoke the python polynomial roots solver.

# output variables
droptime = np.full((numcoef,numover,2),0.) # hold two roots for each condition
finalrotation = np.full((numcoef,numover),0.) # rotation at floor
timetofloor = np.full((numcoef,numover),0.) # drop time to floor

# debug
# print("max coefficient of friction:", coeffric[4])
# print("Max overhang (%): ", overhang[4])
print("numcoef: ", numcoef)
print("numover: ", numover)
# print("sample final rotation: ", finalrotation[1,3])
# print("finalrotation.dtype: ", finalrotation.dtype)
# print("droptime.dtype: ", droptime.dtype)
# print("andvel.dtype: ", angvel.dtype)
````

````python
ndexa = 0
ndexb = 0

while ndexa < numcoef: # only need one loop for now since coeffric and overhang have same length
    # print("ndexa: ", ndexa)
    while ndexb < numover:
        # print("ndexb: ", ndexb) 
        
        # Set first term of each variable
        # time variables
        # count : count is aready completely filled from 0 to numit
        # currenttime[0] is already set to 0
        
        # range of parameters
        # coeffric - hardwired to let the coefficient of friction vary from 0.2 to 1.0 by 0.2
        # overhang - hardwired to let initial overhang vary from .55 to .95 in percentage terms
        
        # general location of cm variables
        rside[ndexa,ndexb,0] = twidth*overhang[ndexb]
        lside[ndexa,ndexb,0] = twidth-rside[ndexa,ndexb,0]
        
        # torque calculation variables
        armr[ndexa,ndexb,0] = rside[ndexa,ndexb,0]/2.0
        arml[ndexa,ndexb,0] = lside[ndexa,ndexb,0]/2.0
        weightr[0] = lindensity*rside[ndexa,ndexb,0]*gravity # weight of overhang
        weightl[0] = lindensity*lside[ndexa,ndexb,0]*gravity # weight over table
        
        # moment of inertia variables
        dparallel[ndexa,ndexb,0] = rside[ndexa,ndexb,0] - twidth/2.  # value changes if slipping
        tinertianow[ndexa,ndexb,0] = tinertiacm + tmass*dparallel[ndexa,ndexb,0]**2 
        # ndexfinal[ndexa,ndexa] is already set to 0
        
        # rotation variables
        # angvel[ndexa,ndexb,0] is already set to 0
        # angpos[ndexa,ndexb,0] is already set to 0
        torqpos[ndexa,ndexb,0] = (overhang[ndexb]*twidth/2)*(tmass*overhang[ndexb]*gravity)
        torqneg[ndexa,ndexb,0] = -((1.0-overhang[ndexb])*twidth/2)*(tmass*(1.0-overhang[ndexb])*gravity)
        torqnet[ndexa,ndexb,0] = torqpos[ndexa,ndexb,0]+torqneg[ndexa,ndexb,0]
        angaccel[ndexa,ndexb,0] = torqnet[ndexa,ndexb,0]/tinertianow[ndexa,ndexb,0]
             
        # slipping variables
        # friction[ndexa,ndexa,0] is already set to 0
        # latgforce[ndexa,ndexa,0] is already set to 0
        # parallelaccel[ndexa,ndexa,0] is already set to 0
        # These arrays had to be added later as I needed to deal with the toast slipping off the edge
        # slipdisplace[ndexa,ndexa,0] = 0.0 # is already set to 0
        slipposx[ndexa,ndexb,0] = rside[ndexa,ndexb,0]- twidth/2.0 # CM relative to pivot due to overhang
        # slipposy[ndexa,ndexa,0]  is already set to 0
        # slipveltot[ndexa,ndexa,0] is already set to 0
        # slipvelx[ndexa,ndexa,0] is already set to 0
        # slipvely[ndexa,ndexa,0] is already set to 0
        
        # kinematic coefficients 
        # quadcoef[ndexa,ndexa,0:2] depend on conditions when toast leaves edge and are set later
        
        # output variables
        # droptime[ndexa,ndexa,0] is already set to 0
        # finalrotation[ndexa,ndexa] is already set to 0
        # timetofloor[ndexa,ndexa] = is already set to 0
        
        ndexb = ndexb + 1
    ndexb = 0
    ndexa = ndexa + 1
# debug
# print("numcoef: ", numcoef)
# print("numover: ", numover)                             
````

### When it starts to slip...(initially ignored)

When the toast starts to slip things get complicated in a hurry. Perhaps most obviously the pivot point starts to move which means all of the torques and moment arms change as well as the moment of intertia. That will all be sort of straightforward. Tracking the motion of the toast as it slides off the edge seems painful since the acceleration and velocity will be in a slightly different direction with each successive iteration. Yikes.....

Remember that python keeps track of loops and other programming features through the indents in the code. All of this part of the model will need to take place inside the 'if-else' conditional test' part way through the calculation. To be more specific it is the 'else' part of the conditional test where all the action has to happen

````python
ndex1 = 0 # cycles through 5 coeffric
ndex2 = 0 # cycles through 5 overhangs
ndex3 = 0 # tracks iterations each time

while (ndex1 < numcoef):
    print("Coefficient of Friction: ", coeffric[ndex1])
    while (ndex2 < numover):
        print("Overhang: ", overhang[ndex2])
        while (ndex3 < numit) and (angpos[ndex1,ndex2,ndex3] < anglimit):
            #print("Iteration: ", ndex3)
    
            # These calculations take place in every iteration regardless of whether it's slipping or not.
            # moment of inertia NOW - ndex1    
            dparallel[ndex1,ndex2,ndex3] = rside[ndex1,ndex2,ndex3] - twidth/2.  # value changes if slipping
            tinertianow[ndex1,ndex2,ndex3] = tinertiacm + tmass*dparallel[ndex1,ndex2,ndex3]**2     
        
            # torqnet NOW - ndex1
            torqpos[ndex1,ndex2,ndex3] = np.cos(angpos[ndex1,ndex2,ndex3])*armr[ndex1,ndex2,ndex3]*weightr[ndex1,ndex2,ndex3]
            torqneg[ndex1,ndex2,ndex3] = -np.cos(angpos[ndex1,ndex2,ndex3])*arml[ndex1,ndex2,ndex3]*weightl[ndex1,ndex2,ndex3]
            torqnet[ndex1,ndex2,ndex3] = torqpos[ndex1,ndex2,ndex3] + torqneg[ndex1,ndex2,ndex3]
               
            # angular acceleration NOW -ndex+1
            angaccel[ndex1,ndex2,ndex3] = torqnet[ndex1,ndex2,ndex3]/tinertianow[ndex1,ndex2,ndex3]   
                
            # NEXT position and velocity after tstep - ndex1+1              
            angvel[ndex1,ndex2,ndex3+1] = angvel[ndex1,ndex2,ndex3] + angaccel[ndex1,ndex2,ndex3]*(tstep/1000.0)
            angpos[ndex1,ndex2,ndex3+1] = angpos[ndex1,ndex2,ndex3] + angvel[ndex1,ndex2,ndex3]*(tstep/1000.0) + 0.5*angaccel[ndex1,ndex2,ndex3]*(tstep/1000.0)*(tstep/1000.0)
            currenttime[ndex3+1] = currenttime[ndex3] + tstep
               
            # determine if the toast is slipping
            # calculate normal, friction, and lateral forces NOW - ndex1
            currentnormal = (weightr[ndex1,ndex2,ndex3] + weightl[ndex1,ndex2,ndex3])*np.cos(angpos[ndex1,ndex2,ndex3])
            friction[ndex1,ndex2,ndex3] = currentnormal*coeffric[ndex1]
            latgforce[ndex1,ndex2,ndex3] = (weightr[ndex1,ndex2,ndex3] + weightl[ndex1,ndex2,ndex3])*np.sin(angpos[ndex1,ndex2,ndex3])                
            parallelaccel[ndex1,ndex2,ndex3] = (latgforce[ndex1,ndex2,ndex3] - friction[ndex1,ndex2,ndex3])/(tmass)
            
            # This is where I have to deal with the toast slipping. When the parallelaccel > 0
            # then the toast is starting to slip. 
            
            if parallelaccel[ndex1,ndex2,ndex3] < 0.0: # NOT slipping
                parallelaccel[ndex1,ndex2,ndex3] = 0.0
                # update variables for next step = ndex1+1
                rside[ndex1,ndex2,ndex3+1] = rside[ndex1,ndex2,ndex3]
                lside[ndex1,ndex2,ndex3+1] = twidth - rside[ndex1,ndex2,ndex3+1]
                armr[ndex1,ndex2,ndex3+1] = rside[ndex1,ndex2,ndex3+1]/2.0
                arml[ndex1,ndex2,ndex3+1] = lside[ndex1,ndex2,ndex3+1]/2.0
                weightr[ndex1,ndex2,ndex3+1] = lindensity*rside[ndex1,ndex2,ndex3+1]*gravity # weight of overhang
                weightl[ndex1,ndex2,ndex3+1] = lindensity*lside[ndex1,ndex2,ndex3+1]*gravity # weight over table
                slipangle[ndex1,ndex2] = angpos[ndex1,ndex2,ndex3+1] # keep updating the slip angle until is starts slipping. 
            
            else:
                print("Toast is slipping!!; ndex3: ", ndex3)
                # determine NEXT sliding velocity - ndex1+1
                slipvelx[ndex1,ndex2,ndex3+1] = slipvelx[ndex1,ndex2,ndex3] + np.cos(angpos[ndex1,ndex2,ndex3])*parallelaccel[ndex1,ndex2,ndex3]*tstep/1000.
                slipvely[ndex1,ndex2,ndex3+1] = slipvely[ndex1,ndex2,ndex3] - np.sin(angpos[ndex1,ndex2,ndex3])*parallelaccel[ndex1,ndex2,ndex3]*tstep/1000.
                slipveltot[ndex1,ndex2,ndex3+1] = np.sqrt(slipvelx[ndex1,ndex2,ndex3+1]**2 + slipvely[ndex1,ndex2,ndex3+1]**2)
                  
                # determine NEXT slid position - ndex1+1
                slipposx[ndex1,ndex2,ndex3+1] = slipposx[ndex1,ndex2,ndex3] + slipvelx[ndex1,ndex2,ndex3+1]*tstep/1000.
                slipposy[ndex1,ndex2,ndex3+1] = slipposy[ndex1,ndex2,ndex3] + slipvely[ndex1,ndex2,ndex3+1]*tstep/1000.
                slipdisplace[ndex1,ndex2,ndex3+1] = np.sqrt(slipposx[ndex1,ndex2,ndex3+1]**2 + slipposy[ndex1,ndex2,ndex3+1]**2)
            
                # find NEXT overhang, this affects the moment of inertia - ndex1+1
                rside[ndex1,ndex2,ndex3+1] = rside[ndex1,ndex2,ndex3]+slipdisplace[ndex1,ndex2,ndex3+1]
                lside[ndex1,ndex2,ndex3+1] = twidth - rside[ndex1,ndex2,ndex3+1]
                weightr[ndex1,ndex2,ndex3+1] = lindensity*rside[ndex1,ndex2,ndex3+1]*gravity # weight of overhang
                weightl[ndex1,ndex2,ndex3+1] = lindensity*lside[ndex1,ndex2,ndex3+1]*gravity # weight over table
                
                # debugging help
                # print("lateral accel (cm/s^2) : ", parallelaccel[ndex1])
                # print("lateral g force: ", latgforce[ndex1])
                # print("currenttime: ", currenttime[ndex1])
                # print("velx: %.3f vely %.3f posx %.3f posy %.3f " % (slipvelx[ndex1],slipvely[ndex1],slipposx[ndex1],slipposy[ndex1]))
                # print("slip velocity %.3f slip displacement %.3f  " % (slipveltot[ndex1],slipdisplace[ndex1]))
                # inputcont = input("continue?")    
                
                       
            # debugging help
            # print("Tpos: %.3f Tneg %.3f Ttot %.3f angaccel %.3f " % (torqpos2[ndex4],torqneg2[ndex4],torqtot[ndex4],angaccel2[ndex4]))
            # print("cos(angpos): ", np.cos(angpos2[ndex2]))
            # print("pos %.3f pos+ %.3f vel %.3f vel+ %.3f accel %.3f " % (angpos2[ndex4],angpos2[ndex4+1],angvel2[ndex4],angvel2[ndex4+1],angaccel2[ndex4]))
            # inputcont = input("continue?")    
            
            # test for end point of rotation
            
            if angpos[ndex1,ndex2,ndex3+1] > (np.pi/2.0):
                ndex3 = ndex3 + 1
                print ("Got to 90 degrees at ndex3: ", ndex3)
                break # get out of the loop
            
            ndex3 = ndex3 + 1 # go to the next time increment
            # inputcont = input("continue to next iteration?")
            
        ndexfinal[ndex1,ndex2] = ndex3
        print("final index: ", ndexfinal[ndex1,ndex2])
        print("Tpos: %.3f Tneg %.3f Ttot %.3f angaccel %.3f : torque 0.0 and angaccel 0.0" % (torqpos[ndex1,ndex2,ndex3],torqneg[ndex1,ndex2,ndex3],torqnet[ndex1,ndex2,ndex3],angaccel[ndex1,ndex2,ndex3]))
        print("pos %.3f vel %.3f : angular position 1.55ish" % (angpos[ndex1,ndex2,ndex3],angvel[ndex1,ndex2,ndex3]))
        print("Angle at which slipping begins is %.3f radians" % (slipangle[ndex1,ndex2]))
            
        ndex3 = 0
        ndex2 = ndex2 + 1
        print("On to next Overhang: ndex2: ", ndex2)
        # inputcont = input("continue to next overhang?") 
    
    ndex2 = 0
    ndex1 = ndex1 + 1
    print("On to next Coefficient of Friction: ndex1: ", ndex1)
    # inputcont = input("continue to next coefficient of friction?")

print("Done with all options:")
                                                                        
````

### New Drop time

To get the rotation of the toast I need to calculate the drop time taking into account that because of slipping (and rotation actually) the toast has some downward velocity when it comes off the edge of the counter. That will slightly reduce the drop time.

````python
# identify the coefficients
ndex7 = 0 # coeffric index
ndex8 = 0 # overhang index

while ndex7 < numcoef: # cycle through coef fric
    # print("coef fric: ", ndex7)
    while ndex8 < numover: # cycle through overhang
        # print("overhang:", ndex8)
        quadcoef[ndex7,ndex8,0] = -gravity/2.0
        quadcoef[ndex7,ndex8,2] = counterheight
        tempindex = ndexfinal[ndex7,ndex8]-1                     
        quadcoef[ndex7,ndex8,1] = slipvely[ndex7,ndex8,tempindex]
        
        # find droptime roots
        droptime[ndex7,ndex8,] = np.roots(quadcoef[ndex7,ndex8,])
        
        if droptime[ndex7,ndex8,0] > 0.0: # assume 2 roots and only one is positive....could be a problem
            finalrotation[ndex7,ndex8] = droptime[ndex7,ndex8,0]*angvel[ndex7,ndex8,ndexfinal[ndex7,ndex8]]
            timetofloor[ndex7,ndex8] = droptime[ndex7,ndex8,0]
        else:
            # finalrotation[ndex7,ndex8] = droptime[ndex7,ndex8,1]*angvel[ndex7,ndex8,ndexfinal[ndex7,ndex8]]
            finalrotation[ndex7,ndex8] = droptime[ndex7,ndex8,1]*angvel[ndex7,ndex8,ndexfinal[ndex7,ndex8]]
            timetofloor[ndex7,ndex8] = droptime[ndex7,ndex8,1]
        
        print("Final Report:")
        print("Final Rotation at Floor (rad): ", finalrotation[ndex7,ndex8])
        print("Angular velocity coming off the table (rad/s):", angvel[ndex7,ndex8,ndexfinal[ndex7,ndex8]])
        print("Time to reach floor (s):", timetofloor[ndex7,ndex8])
        print("Initial overhang (%):", overhang[ndex8])
        print("Coefficient of Friction:", coeffric[ndex7])
        print("Angle at which slipping started (rad):", slipangle[ndex7,ndex8])
        print("Time until comes off edge (ms): ", currenttime[ndexfinal[ndex7,ndex8]])
        
        # debug
        # print("coef 0 (t^2): ", quadcoef[0])
        # print("coef 1 (t): ", quadcoef[1])
        # print("coef 2 (const):", quadcoef[2])
        # print("root 1: ", droptime[0])
        # print("root 2: ", droptime[1])
        
        print("next overhang:")
        # inputcont = input("continue to next overhang?")
        ndex8 = ndex8 + 1
    
    ndex8 = 0
    print("next coef friction:")
    # inputcont = input("continue to next coefficient of friction?")
    ndex7 = ndex7 + 1
    
print("Done with all drop times and rotation calculations.")
                                                
````
